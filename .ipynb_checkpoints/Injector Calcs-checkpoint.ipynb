# Injector Calculations

## Introduction
The oxidiser that shall be used for SFII, nitrous oxide ($N_{2}O$), boils at ambient temperatures. This causes the tank to have a pressure set by the liquid-vapor thermal equilibrium that for a given tank temperature. Because of this, $N_2O$ is often described as 'self-pressurising'. 

For injector design, its necessary to appreciate the two phase nature of $N_2O$ for orriface sizing. Here we will concentrate on common models for injector sizing which can be applied to the SFIII injector design.

Most of the analysis is lifted from the papers by [Waxman and Zimmerman](https://ntrs.nasa.gov/api/citations/20190001326/downloads/20190001326.pdf) and [Dyer](https://www.researchgate.net/profile/Jonathan-Dyer-5/publication/268482381_Modeling_Feed_System_Flow_Physics_for_Self-Pressurizing_Propellants/links/64a822d3c41fb852dd5b4d6e/Modeling-Feed-System-Flow-Physics-for-Self-Pressurizing-Propellants.pdf?_tp=eyJjb250ZXh0Ijp7ImZpcnN0UGFnZSI6InB1YmxpY2F0aW9uIiwicGFnZSI6InB1YmxpY2F0aW9uIn19). Also check [Solomon's paper](https://digitalcommons.usu.edu/cgi/viewcontent.cgi?article=1110&context=gradreports) for more detail about Dyer's NHNE model.

## Single-Phase Incompressible (SPI) Liquid Flow:
**key assumptions:**
* **Liquid only:** no vapor phase
* **Incompressible:** density is constant

It is an adequate assumption to assume that most liquid propellants are incompressible since the density of the liquid is much greater than the density of the vapor phase.

$$
\dot{m}_{SPI} = Q \rho_{l} = C_d A \sqrt{ 
\frac
{2 \rho \Delta P}
{1 - \left( \frac{A}{A_1} \right)^2}
}
$$

where:
* $\dot{m}_{SPI}$ is the mass flow rate of the liquid
* $Q$ is the volumetric flow rate of the liquid
* $\rho_{l}$ is the density of the liquid
* $C_d$ is the discharge coefficient
* $A$ is the area of the injector orifice
* $A_1$ is the upstream cross-sectional area
* $\Delta P$ is the pressure drop across the injector

## Homogeneous Equilibrium (HEM) Model 
The homogeneous equilibrium model assumes that the liquid and vapour phases are in thermal equilibrium at the injector exit. 

**key assumptions:**
* Liquid-vapor thermal equilibrium at the injector exit
* Isentropic flow across injector

Following these assumptions, the mass flow rate of the injector can be calculated as follows: 

$$
\dot{m} = C_{d}A_{\rho_{2}} \sqrt{2(h_1-h_2)}
$$
    
where :
* $C_{d}$ is the discharge coefficient
* $A_{\rho_{2}}$ is the area of the injector orifice
* $h_1$ is the upstream enthalpy
* $h_2$ is the downstream enthalpy

**Limitations**
* This model may be appropriate to predict mass flow only up to the downstream critical pressure  
* Equilibrium may not be achieved for short tube injectors as shown in Dyer's paper

### Two-Phase Flow with Non-Equilibrium Effects
Dyer goes on to propose that non-equilibrium effects can be accounted for by:

1) **Super-heating of the liquid phase:** this occurs during expansion 
2) **Finite vapor bubble growth rates**

#### Characteristic Bubble Growth Rate, $\tau_r$
To account for the finite bubble growth rate, Dyer proposes a characteristic bubble growth rate, $\tau_b$ which is defined as:
$$
\tau_b \equiv \sqrt{ \frac{3}{2} 
\frac{\rho_{l}}{P_v - P_2}}
$$

where:
* $\rho_{l}$ is the liquid density
* $P_v$ is the vapour pressure
* $P_2$ is the downstream pressure

#### Fluid Residence Time, $\tau_r$
The fluid residence time is defined as the time it takes for the fluid to travel from the injector face to the downstream pressure. This is given by:

$$
\tau_r \equiv \frac{L}{u} = 
\frac{L}{\sqrt{ \frac{2 \nabla P}{\rho_{l} }}} =
L \sqrt{ \frac{\rho_{l}}{2 \nabla P}}
$$

where:
* $L$ is the injector length
* $u$ is the velocity of the fluid
* $\nabla P$ is the pressure drop across the injector

### Non-Homogeneous Non-Equilibrium (NHNE) Model
Coined by Solomon, the NHNE model is a combination of the homogeneous equilibrium model and the two-phase flow with non-equilibrium effects. The NHNE model uses the non-equilibrium parameter, $\kappa$ to account for the non-equilibrium effects.


$$
\kappa = \sqrt{ \frac{P_1 - P_2}{P_v - P_2}} \propto \frac{\tau_r}{\tau_b}
$$

The mass flow rate can then be calculated as follows:

$$
\dot{m} = A \left(
\frac{\kappa}{1 + \kappa} \dot{m}_{SPI} +
\frac{1}{1 + \kappa} \dot{m}_{HEM}
\right)
$$


## Retrieving Thermophysical Properties of $N_2O$
The thermophysical properties of $N_2O$ are retrieved from the Aspen Plus database. These have been saved in the spreadsheet `Aspen Data.xlsx`.

Enthalpy and density are retireved for both liquid and vapour phases. for a range of suitable pressures and temperatures. The data is then interpolated to give a continuous function for the thermophysical properties.



```python
from scipy.interpolate import griddata
import numpy as np
import pandas as pd

# Specify the path to the data
file_path = 'Aspen Data.xlsx'

# Load the data
liquid_data = pd.read_excel(file_path, sheet_name='LIQUID')
vapor_data = pd.read_excel(file_path, sheet_name='VAPOR')

def interpolate_property(T, P, phase, prop):
    """
    Interpolates the property value for a given temperature, pressure, and phase.

    Args:
    T (float): Temperature in K.
    P (float): Pressure in bar.
    phase (str): 'liquid' or 'vapor'.
    prop (str): 'density' or 'enthalpy'.

    Returns:
    float: Interpolated property value.
    """
    # Select the appropriate dataset
    if phase.lower() == 'liquid':
        data = liquid_data
    elif phase.lower() == 'vapor':
        data = vapor_data
    else:
        raise ValueError("Phase must be 'liquid' or 'vapor'")

    # Check if the property is valid
    if prop.lower() not in ['density', 'enthalpy']:
        raise ValueError("Property must be 'density' or 'enthalpy'")

    # Extract the required columns
    points = data[['Temperature, K', 'Pressure, bar']].values
    values = data[prop.capitalize() + ', kg/m3' if prop.lower() == 'density' else prop.capitalize() + ', kJ/kg'].values
    
    # Perform interpolation
    interpolated_value = griddata(points, values, ([T, P]), method='linear')

    # Handle cases where interpolation is not possible (e.g., out of range)
    if interpolated_value[0] is np.nan:
        return "Interpolation not possible for given T and P in this phase."

    return interpolated_value[0]
```

Testing the interpolation function:

```python
print(f'Interpolated density of liquid N2O at 281 K and 43 bar: {interpolate_property(291, 43, "liquid", "density")} kg/m3')
print(f'Interpolated enthalpy of vapor N2O at 360 K and 20 bar: {interpolate_property(300, 20, "vapor", "enthalpy")} kJ/kg')
```

## Define functions
define the functions for INC and HEM model

```python
def area(d):
    """Calculate the cross-sectional area of a pipe given its diameter."""
    return np.pi * d ** 2 / 4

def INC(A, rho1, P1, P2, Cd=0.66):
    """Calculate the mass flow rate using the Inertial Control Node method."""
    return Cd * A * np.sqrt(2 * rho1 * (P1 - P2))

def HEM(A, rho2, h2, h1, Cd=0.66):
    """Calculate the mass flow rate using the Homogeneous Equilibrium Model."""
    return Cd * A * rho2 * (np.sqrt(2 * (h2 - h1)))
```

## Plot the data

```python
import pandas as pd
import numpy as np
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load the data from the Excel file
file_path = 'Aspen Data.xlsx'  # Adjust the path as needed
liquid_data = pd.read_excel(file_path, sheet_name='LIQUID')
vapor_data = pd.read_excel(file_path, sheet_name='VAPOR')

def interpolate_property(data, T, P, prop):
    """Interpolates the property value for a given temperature, pressure, and property."""
    # Extract the required columns
    points = data[['Temperature, K', 'Pressure, bar']].values
    values = data[prop.capitalize() + ', kg/m3' if prop.lower() == 'density' else prop.capitalize() + ', kJ/kg'].values
    
    # Perform interpolation
    interpolated_value = griddata(points, values, ([T, P]), method='linear')

    # Handle cases where interpolation is not possible
    if np.isnan(interpolated_value[0]):
        return np.nan  # Return NaN if interpolation is not possible

    return interpolated_value[0]
```

```python
def area(d):
    """Calculate the cross-sectional area of a pipe given its diameter."""
    return np.pi * d ** 2 / 4

def INC(A, rho1, P1, P2, Cd=0.66):
    """Calculate the mass flow rate using the Inertial Control Node method."""
    return Cd * A * np.sqrt(2 * rho1 * (P1 - P2))

def HEM(A, rho2, h2, h1, Cd=0.66):
    """Calculate the mass flow rate using the Homogeneous Equilibrium Model."""
    delta_h = h2 - h1
    if delta_h < 0:  # Check if delta_h is negative
        return 0  # Return 0 or any other appropriate value
    else:
        return Cd * A * rho2 * np.sqrt(2 * delta_h)
```

```python
# Define the range of diameters and pressures
P1_values = np.linspace(30, 65, 100)[:, np.newaxis]  # Pressures from 30 to 65 bar
d_values = np.linspace(0, 2, 100)[np.newaxis, :]  # Diameters from 0 to 2.5 m
d_values /= 1e3  # Convert from millimeters to meters

# Compute the properties and mINC values
mINC_2D = np.zeros((len(P1_values), len(d_values[0])))
mHEM_2D = np.zeros((len(P1_values), len(d_values[0])))


P2 = 20     # bar (constant)
T1 = 285    # K (constant)

for i, P1 in enumerate(P1_values.flatten()):
    rho1 = interpolate_property(liquid_data, T1, P1, 'density')
    rho2 = interpolate_property(liquid_data, T1, P2, 'density')
    h1 = interpolate_property(liquid_data, T1, P1, 'enthalpy')
    h2 = interpolate_property(vapor_data, T1, P2, 'enthalpy')
    for j, d in enumerate(d_values.flatten()):
        A_j = area(d)
        mINC = INC(A_j, rho1, P1*1e5, P2*1e5)
        mINC_2D[i, j] = mINC
        mHEM = HEM(A_j, rho1, h1, h2)
        mHEM_2D[i, j] = mHEM

# Multiply the mINC values by 1000
mINC_1000x = mINC_2D * 1e3

# Convert diameter values from meters to millimeters
d_values_mm = d_values[0] * 1000  # Convert from meters to millimeters

# Select sensible tick marks for both axes
d_ticks = np.round(d_values_mm[::10], 1)
P1_ticks = np.round(P1_values[::10, 0], 1)
```

```python
# Setting up the heatmap plot
plt.figure(figsize=(12, 9))
sns.heatmap(
    mINC_2D,
    xticklabels=d_ticks,
    yticklabels=P1_ticks,
    annot=False,
    fmt='.2f',
    cmap='inferno',
    cbar_kws={'label': 'INC mass flow rate'}
)

# Adding contour lines
contour = plt.contour(mINC_2D, colors='white', linestyles='solid', linewidths=2, levels=15)
plt.clabel(contour, inline=True, fontsize=12, fmt='%.2f')

# Adding labels and title
plt.xlabel('Diameter (mm)')
plt.ylabel('Upstream pressure (bar)')
plt.title('INC Mass Flow Rate per Orrifice Diameter', fontsize=16)

# Adjusting the ticks on the x and y axes
plt.xticks(np.arange(0, len(d_values_mm), 10), d_ticks)
plt.yticks(np.arange(0, len(P1_values), 10), P1_ticks)
plt.gca().invert_yaxis()

# Display the heatmap
plt.show()
```

```python
from ipywidgets import interact, IntSlider

def plot_with_orifices(n_orifices):
    # Calculate the total mass flow rate
    total_mINC_2D = n_orifices * mINC_2D

    # Setting up the heatmap plot
    plt.figure()
    sns.heatmap(
        total_mINC_2D,
        xticklabels=d_ticks,
        yticklabels=P1_ticks,
        annot=False,
        fmt='.2f',
        cmap='inferno',
        cbar_kws={'label': 'Total INC mass flow rate'}
    )

    # Adding contour lines
    contour = plt.contour(total_mINC_2D, colors='white', linestyles='solid', linewidths=2, levels=15)
    plt.clabel(contour, inline=True, fontsize=12, fmt='%.2f')

    # Adding labels and title
    plt.xlabel('Diameter (mm)')
    plt.ylabel('Upstream pressure (bar)')
    plt.title(f'Total INC Mass Flow Rate for {n_orifices} Orrifices', fontsize=16)

    # Adjusting the ticks on the x and y axes
    plt.xticks(np.arange(0, len(d_values_mm), 10), d_ticks)
    plt.yticks(np.arange(0, len(P1_values), 10), P1_ticks)
    plt.gca().invert_yaxis()

    # Display the heatmap
    plt.show()

# Create a slider for the number of orifices
interact(plot_with_orifices, n_orifices=IntSlider(min=1, max=20, step=1, value=1));
```
